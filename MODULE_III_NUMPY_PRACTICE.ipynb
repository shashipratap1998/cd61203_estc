```python
# INTEGRALS AND DERIVATIVES
x = np.linspace(1, 10, 100)
y = 1/x**2 * np.sin(x)
# derivative
dydx = np.gradient(y, x) # dydx is variable
# integral
y_int = np.cumsum(y) * (x[2] - x[0]) # cumulative sum = integral
# (x[1] - x[0]) is like the 'dx' part in normal integration wrt x!
# The differenc is just that we are taking the discrete not continous
# So, (x[1] - x[0]) represents x.
```

```python
np.cumsum(np.array([1, 2, 3 , 4]))
# Example of cumulative sum
```

```python
plt.plot(x,y, label = 'Curve 1')
plt.plot(x, dydx, label = 'Curve 2')
plt.plot(x, y_int, label = 'Curve 3')
plt.legend()
```

## Question 4:

The power emitted by a blackbody is $P = A \sigma \epsilon T^4$. After measuring the temperature of a star you find that $T(t) = T_0 \frac{1}{1+e^{-kt}}$. Plot the total energy emitted by the star as a function of time using the fact that $E(t) = \int_{t'=0}^{t'=t} P(t') dt' $

```python
kt = np.linspace(0, 10, 1000)
P = (1/(1+np.exp(-kt)))**4 # Here P = P/AsigmaT0^4
E = np.cumsum(P) * (kt[1]-kt[0])
```

```python
plt.plot(kt,E)
plt.xlabel('$kt$', fontsize=20)
plt.ylabel(r'$\left(\frac{k}{A \sigma \epsilon T_0^4}\right) E(kt)$', fontsize=20)
plt.show()
```

## Question 4 (tougher)

Find the solutions to $\left(\frac{d^2}{dx^2} + (10x)^2\right) f = \lambda f$ with boundary conditions $f(0)=f(1)=0$

* use the eigenvalue method and the fact that $\frac{d^2 f}{dx^2} \approx \frac{f_{i+1} + f_{i-1} - 2f_i}{\Delta x^2}$ where $f_i = f(i \Delta x)$ and $i$ is an integer.

```python
N = 1000
x = np.linspace(0, 1, N+1)
dx = x[1]-x[0]
dx
```

```python
main_diag = -2* np.ones(N-1) #N-1 because we don't want end points i.e, 0 and N+1; So there will be N elements with N-1 interval
# N+1 - 2 = N-1
main_diag
```

```python
off_diag = np.ones(N-2)
off_diag
```

```python
# diagonalise the main element that we have prepared
diag_example = np.diag(main_diag)
diag_example
```

```python
# now add the 'ones' that we need in the same matrix
derivative_matrix = (np.diag(main_diag) + np.diag(off_diag, -1) + np.diag(off_diag, 1))/dx**2
derivative_matrix
```

```python
x2_matrix = np.diag(10*x[1:-1]**2)
x2_matrix
```

```python
LHS_matrix = derivative_matrix + x2_matrix
LHS_matrix
```

```python
w, v = np.linalg.eigh(LHS_matrix)
w
```

```python
plt.plot(-1*v[:,-1])
plt.plot(-1*v[:,-2])
plt.plot(-1*v[:,-3])
plt.plot(-1*v[:,-4])
```
