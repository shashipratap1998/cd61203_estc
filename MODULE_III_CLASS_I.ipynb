```python
# Function Checking type of arguments-recursive
def factorial (n):
  if not isinstance(n, int):
    print("Factorial is only defined for integers only!")
    return None
  elif n < 0:
    print("Factorial is not valid for negative numbers")
    return None
  elif n == 0:
    return 1
  else:
    return n * factorial(n-1) #recursively


factorial(5)
```

```python
# Default arguments/Function overloading
def func1(arg1, arg2=None):
  if arg2 is None:
    print(arg1)
  else:
    print(arg1, arg2)

#func1('hello')
func1('hello', 'world')
```
