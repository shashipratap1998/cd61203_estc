```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the function ax^2
def f(x, a):
    return a * x**2

# Constants
a = 1  # Value of a (you can change this as needed)

# Integration limits
x_lower = -np.pi
x_upper = np.pi

# Perform the integration
val, abserr = quad(f, x_lower, x_upper, args=(a,))
print("Integral value =", val, ", Absolute error =", abserr)

# Generate x values
n = 256
X = np.linspace(x_lower, x_upper, n, endpoint=True)

# Evaluate the function f(x) = ax^2 for each x
Y = f(X, a)

# Line styles: solid, dashed, and dotted
line_styles = ['solid', 'dashed', 'dotted']

# Plot the function with customized line width, color, and line style
for i, line_style in enumerate(line_styles):
    plt.plot(X, Y, label=f'{a}x^2 ({line_style})', linewidth=2, linestyle=line_style)

# Add a shaded region under the curve
plt.fill_between(X, 0, Y, where=(X >= x_lower) & (X <= x_upper), color='lightgray', alpha=0.5)

plt.xlabel('x')
plt.ylabel(f'{a}x^2')
plt.title(f'Plot of {a}x^2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn
from scipy.integrate import quad

# Define the functions for Bessel integrands
def integrand_j0(x):
    return jn(0, x)

def integrand_j3(x):
    return jn(3, x)

# Define the integration limits
x_lower = 0
x_upper = 10

# Perform the integrations
integral_j0, _ = quad(integrand_j0, x_lower, x_upper)
integral_j3, _ = quad(integrand_j3, x_lower, x_upper)

print("Integral of J0(x) from 0 to 10:", integral_j0)
print("Integral of J3(x) from 0 to 10:", integral_j3)

# Generate x values for plotting
x = np.linspace(x_lower, x_upper, 1000)

# Calculate the Bessel functions (J0 and J3) for the given x values
j0_values = jn(0, x)
j3_values = jn(3, x)

# Plot the Bessel functions (J0 and J3)
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(x, j0_values, label='J0(x)')
plt.xlabel('x')
plt.ylabel('J0(x)')
plt.title('Bessel Function J0(x)')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(x, j3_values, label='J3(x)')
plt.xlabel('x')
plt.ylabel('J3(x)')
plt.title('Bessel Function J3(x)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import yn
from scipy.integrate import quad

# Define the functions for Bessel integrands (second kind)
def integrand_y0(x):
    return yn(0, x)

def integrand_y2(x):
    return yn(2, x)

# Define the integration limits
x_lower_y0 = 0
x_upper_y0 = 3

x_lower_y2 = 0
x_upper_y2 = 2

# Perform the integrations
integral_y0, _ = quad(integrand_y0, x_lower_y0, x_upper_y0)
integral_y2, _ = quad(integrand_y2, x_lower_y2, x_upper_y2)

print("Integral of Y0(x) from 0 to 3:", integral_y0)
print("Integral of Y2(x) from 0 to 2:", integral_y2)

# Generate x values for plotting
x_y0 = np.linspace(x_lower_y0, x_upper_y0, 1000)
x_y2 = np.linspace(x_lower_y2, x_upper_y2, 1000)

# Calculate the Bessel functions (Y0 and Y2) for the given x values
y0_values = yn(0, x_y0)
y2_values = yn(2, x_y2)

# Plot the Bessel functions (Y0 and Y2)
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(x_y0, y0_values, label='Y0(x)')
plt.xlabel('x')
plt.ylabel('Y0(x)')
plt.title('Bessel Function Y0(x)')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(x_y2, y2_values, label='Y2(x)')
plt.xlabel('x')
plt.ylabel('Y2(x)')
plt.title('Bessel Function Y2(x)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import statistics
from scipy.linalg import solve
from matplotlib import pyplot as plt

t_max = 5;
n_max = 31;
#np = n_max;
h = 10;
g = -9.81;
v_ini=0;

t_domain = np.linspace(0, t_max, n_max)
res = [t_domain[i + 1] - t_domain[i] for i in range(len(t_domain)-1)]
dt = statistics.mean(res)

#rhs side
b=np.zeros((n_max,1))

b=(g*dt*dt)+b
b[0,0]=h
b[1,0]=v_ini*dt


#lhs side
A=np.zeros((n_max,n_max))
for i in range(2,n_max):
  A[i,i-2]=1;
  A[i,i-1]=-2;
  A[i,i]=1;

A[0,0]=1;
A[1,0]=-1;
A[1,1]=1;


X = np.linalg.inv(A).dot(b)
X_A=h+0.5*g*(np.square(t_domain))+v_ini*t_domain

print(X)
print(X_A)

plt.rcParams["figure.figsize"] = [5,3.8]
plt.rcParams["figure.autolayout"] = True
plt.rcParams['axes.linewidth'] = 1
plt.rc('xtick',labelsize=14)
plt.rc('ytick',labelsize=14)

h1, =plt.plot(t_domain,X_A,'r-',linewidth=2,label="analytical")
h2, =plt.plot(t_domain,X,'bo',linewidth=2,label="numerical")

plt.xlabel('t (s)',fontsize=20)
plt.ylabel('x (mm)',fontsize=20)

plt.xlim(0,t_max)
plt.ylim(0,1.1*h)
plt.margins(x=0, y=0)

#
first_legend = plt.legend(handles=[h1,h2], loc='lower right')
plt.show()
```

Spring-mass-damper with forcing

```python
# Simulation of Mass-Spring-Damper System
import numpy as np
import matplotlib.pyplot as plt


# Model Parameters
c = 4 # Damping constant
k = 2 # Stiffness of the spring
m = 20 # Mass
F = 5 # Force


# Simulation Parameters
Ts = 0.1
Tstart = 0
Tstop = 60


N = int((Tstop-Tstart)/Ts) # Simulation length
x1 = np.zeros(N+2)
x2 = np.zeros(N+2)


x1[0] = 0 # Initial Position
x2[0] = 0 # Initial Speed


a11 = 1
a12 = Ts
a21 = -(Ts*k)/m
a22 = 1 - (Ts*c)/m
b1 = 0
b2 = Ts/m


# Simulation
for k in range(N+1):
  x1[k+1] = a11 * x1[k] + a12 * x2[k] + b1 * F
  x2[k+1] = a21 * x1[k] + a22 * x2[k] + b2 * F

# Plot the Simulation Results
t = np.arange(Tstart,Tstop+2*Ts,Ts)


#plt.plot(t, x1, t, x2)
plt.plot(t,x1)
plt.plot(t,x2)
plt.title('Simulation of Mass-Spring-Damper System')
plt.xlabel('t [s]')
plt.ylabel('x(t)')
plt.grid()
plt.legend(["x1", "x2"])
plt.show()
```
