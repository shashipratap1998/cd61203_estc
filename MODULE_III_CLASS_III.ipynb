```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
L = 1  # Length of the square surface
N = 5  # Number of grid points in each direction (including boundaries)
dx = L / (N - 1)  # Grid spacing

# Initialize grid and boundary conditions
T = np.ones((N, N)) * 10  # Initial guess, all points at 10°C

# Set boundary conditions
T[:, 0] = 10  # Bottom
T[:, -1] = 10  # Top
T[0, :] = 100  # Left side
T[-1, :] = 50  # Right side

# Define the source term
def source(x, y):
    return np.sin(x + y)

# Define a function to perform one iteration of the solver
def solve_iteration(T, dx):
    T_new = np.copy(T)
    for i in range(1, N-1):
        for j in range(1, N-1):
            T_new[i, j] = 0.25 * (T[i-1, j] + T[i+1, j] + T[i, j-1] + T[i, j+1] - dx**2 * source(i*dx, j*dx))
    return T_new

# Perform iterations
num_iterations = 1000
for _ in range(num_iterations):
    T = solve_iteration(T, dx)

# Print temperature at specific points
x_index = int(0.25 / dx)
y_index = int(0.25 / dx)
print(f"Temperature at (0.25, 0.25): {T[x_index, y_index]}°C")

x_index = int(0.75 / dx)
y_index = int(0.75/ dx)
print(f"Temperature at (0.75, 0.75): {T[x_index, y_index]}°C")

# Generate coordinates
x = np.linspace(0, L, N)
y = np.linspace(0, L, N)

# Create a meshgrid for contour plot
X, Y = np.meshgrid(x, y)

# Plot the contour
plt.contourf(Y,X, T, cmap='jet', levels=20)
plt.colorbar(label='Temperature (°C)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Temperature Distribution')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Grid parameters
n = 5  # Number of grid points along each axis (assuming it's a square grid)
dx = dy = 1.0 / (n - 1)  # Grid spacing

# Initialize the temperature matrix with boundary conditions
T = np.zeros((n, n))
T[0, :] = 100  # Top boundary
T[n-1, :] = 50  # Bottom boundary
T[1:n-1, 0] = 10  # Left boundary
T[1:n-1, n-1] = 10  # Right boundary

# Numerical approximation using finite difference method
for i in range(1, n - 1):
    for j in range(1, n - 1):
        T[i, j] = (dx**2 * dy**2 * np.sin(i*dx + j*dy) +
                   T[i+1, j] + T[i-1, j] + T[i, j+1] + T[i, j-1]) / (2 * (dx**2 + dy**2))

# Print the temperature at remaining grid points
print("Temperature at remaining grid points:")
for i in range(n):
    for j in range(n):
        if T[i, j] == 0:
            print(f"T({i+1}, {j+1}) = {T[i, j]:.2f} degrees Celsius")

# Create a color contour plot
x = np.linspace(0, 1, n)
y = np.linspace(0, 1, n)
X, Y = np.meshgrid(x, y)

plt.contourf(X, Y, T, cmap='hot')
plt.colorbar(label='Temperature (°C)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Temperature Distribution')
plt.show()
```
