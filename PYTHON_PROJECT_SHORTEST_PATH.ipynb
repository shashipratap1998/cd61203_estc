```python
#The first thrree are important packages require dfo this project
import cv2#this is a library used for image related work
import numpy as np#numoy used for efficient operations
import threading#is amodule

######
#THIS PROGRAM CONTAINS BOTH DFS AND BFS
#TO USE DFS JUST COMMENT BFS AND VICEVERSA AT THE END OF PROGRAM


#CREATED A CLASS POINT AND USED OPERATOR OVERLOADING
class Point(object):#stores the coordinates of teh point clicked on screen

    def __init__(self, x=0, y=0):#this is constructor#fatre clicking x,y will become the actual values
        self.x = x#self. refers to current object
        self.y = y

    def __add__(self, other):#if one piint is (2,4) other is (1,2) then output will be (3,6)
        #print(self.x + other.x, self.y + other.y)
        return Point(self.x + other.x, self.y + other.y)

    def __eq__(self, other):#check if two points are equal or not
        return self.x == other.x and self.y == other.y#return true if both points are same


#initialize the variables
p=0#initialize p
c=0#initialize c
start = Point()#create an object start of Point
end = Point()#create en object end of Point
dir4 = [Point(1,0 ), Point(0, 1), Point(0,-1), Point(-1, 0)]#ffour directions  corresponding t0
#Right,up,down,left

#BFS:TO USE BFS COMMENT DFS AT THE END OF PROGRAM
def BFS(s, e):#s->start, e-->end
    global img, h, w,c#there are some global variables
    solved = False
    queue= []#queeu data structure is used
    visited = [[0 for j in range(w)] for i in range(h)]#if the point is visited or not
    parent = [[Point() for j in range(w)] for i in range(h)]#intialise Point object for each cell in 2-D grid
    queue.append(s)#push start into the queue
    visited[s.y][s.x] = 1#set visited of start to 1
    while len(queue) > 0:#jab tak queue empty nahi hota tab tak chal raha hai
        p = queue.pop(0)#take out the first elemnet of queue
        for d in dir4:#iterate over all directions left ,right,up,down
            cell = p + d#go to the new cell
            #This is for checking the validity of a point
            if (cell.x >= 0 and cell.x < w and cell.y >= 0 and cell.y < h and visited[cell.y][cell.x] == 0 and
                    (img[cell.y][cell.x][0] != 0 or img[cell.y][cell.x][1] != 0 or img[cell.y][cell.x][2] != 0)):#black nahi hona chahiye
                    #there are three channels in an image RBG
                queue.append(cell)#insert in queue
                visited[cell.y][cell.x] = 1#visited ko true kardenge
                img[cell.y][cell.x] = (45,43,78)#assign a random color
                parent[cell.y][cell.x] = p#make the parent of the currnet cell to p
                if cell == e:#kya hum end pe pohunch gaye hain
                    solved = True#seolve ho gaya
                    del queue[:]#queue ko dlete kar rahe hain
                    break#exit the for loop
    #BFS complete hai we have the path
    path = []#store the actual path
    if solved:#solved True
        p = e
        while p != s:#until we get start
            path.append(p)#append parent to path
            p = parent[p.y][p.x]#make the parent to the parant of the currnet cell

        path.append(p)#appending start
        path.reverse()#reverse the path

        for p in path:
            img[p.y][p.x] = [255, 255, 255]#initiale the path to white
            c=c+1
        print("BFS")
        print("Path Found")
        print("Path length is:",c)
    else:
        print("Path Not Found")

#DFS:TO USE DFS COMMENT BFS AT THE END OF PROGRAM
def DFS(s, e):
    global img, h, w,c

    solved = False
    q = []#list
    v = [[0 for j in range(w)] for i in range(h)]#distance array of size w*h all zeros
    parent = [[Point() for j in range(w)] for i in range(h)]#create defaulr Point object
    q.append(s)#insert start
    v[s.y][s.x] = 1#visited start=true
    while len(q) > 0:
        p = q.pop()
        for d in dir4:
            cell = p + d
            if (cell.x >= 0 and cell.x < w and cell.y >= 0 and cell.y < h and v[cell.y][cell.x] == 0 and
                    (img[cell.y][cell.x][0] != 0 or img[cell.y][cell.x][1] != 0 or img[cell.y][cell.x][2] != 0)):
                q.append(cell)
                v[cell.y][cell.x] = v[p.y][p.x] + 1  #child ka distance-parentkadistance +1

                img[cell.y][cell.x] =(30,60,120)
                parent[cell.y][cell.x] = p
                if cell == e:
                    solved = True
                    del q[:]
                    break
    path = []
    if solved:
        p = e
        while p != s:
            path.append(p)
            p = parent[p.y][p.x]
        path.append(p)
        path.reverse()

        for p in path:
            img[p.y][p.x] = [255, 255, 255]
            c+=1
        print("DFS")
        print("Path Found")
        print("Path Length:",c)
    else:
        print("Path Not Found")


def mouse_event(event, pX, pY, flags, param):#function to take in mouse point where clicked
    global img, start, end, p

    if event == cv2.EVENT_LBUTTONUP:
        if p == 0:
            cv2.circle(img,(pX,pY),2,(0,0,255),2)
            start = Point(pX, pY)
            print("start = ", start.x, start.y)
            p += 1
        elif p == 1:
            cv2.circle(img, (pX, pY), 2, (0, 255, 0),2)
            end = Point(pX, pY)
            print("end = ", end.x, end.y)
            p += 1


def disp():#dispalyu the image
    global img
    cv2.imshow("Image", img)
    cv2.setMouseCallback('Image', mouse_event)
    while True:
        cv2.imshow("Image", img)
        cv2.waitKey(1)


img = cv2.imread("maze3.png", cv2.IMREAD_GRAYSCALE)#image ko khola
_, img = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)
img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
h, w = img.shape[:2]#height aur width
print("Select start and end points : ")
t = threading.Thread(target=disp, args=())
t.start()
while p < 2:
    pass


#COMMENT THE TYPE OF METHOD YOU DONT WANT TO SEE
#CURRENTLY IT DISPLAYS BFS ON THE SCREEN
BFS(start, end)
# DFS(start, end)

```
